```python
import pandas as pd
```

```python
def read_space_delimited_csv(file_path):
    """
    Reads a space-delimited CSV file into a pandas DataFrame.

    :param file_path: Path to the space-delimited CSV file.
    :return: pandas DataFrame containing the file data.
    """
    df = pd.read_csv(file_path, delim_whitespace=True)
    return df
```

```python
def read_tab_delimited_txt(file_path):
    """
    Reads a tab-delimited text file with two columns:
    First column becomes new header, second column becomes new first row.

    :param file_path: Path to the tab-delimited text file.
    :return: Tuple of (new_header, new_first_row).
    """
    df = pd.read_csv(file_path, sep='\t', header=None)
    new_header = df.iloc[:, 0].tolist()
    new_first_row = df.iloc[:, 1].tolist()
    return new_header, new_first_row
```

```python
times, g10 = read_tab_delimited_txt("C:/Users/murray98/Documents/Bruce Buffett model/code/dipole_3000_resampled_Amax1.05_Amin-0.6_R0.0036_10kyr.txt")
```

```python
def combine_with_txt_header_and_row(csv_df, new_header, new_first_row):
    """
    Combines the original CSV data with the new header and first row.

    :param csv_df: Original DataFrame with no header.
    :param new_header: List of new column headers.
    :param new_first_row: List of values for the new first row.
    :return: Modified DataFrame.
    """
    num_cols = csv_df.shape[1]
    new_header = new_header[:num_cols]
    new_first_row = new_first_row[:num_cols]

    # Replace column names
    csv_df.columns = new_header

    # Insert new first row
    new_first_df = pd.DataFrame([new_first_row], columns=new_header)
    combined_df = pd.concat([new_first_df, csv_df], ignore_index=True)

    combined_df = combined_df.apply(pd.to_numeric, errors='ignore')
    combined_df = combined_df.multiply(1000) #Convert from microT to nanoT

    return combined_df
```

```python
combined =  combine_with_txt_header_and_row(bb18, times, g10)
```

```python
def save_dataframe_to_csv(df, output_path):
    """
    Saves a pandas DataFrame to a CSV file.

    :param df: pandas DataFrame to save.
    :param output_path: Path to the output CSV file.
    """
    df.to_csv(output_path, index=False)
```
