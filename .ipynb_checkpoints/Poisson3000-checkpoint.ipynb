One-dimensional Poisson model for the geomagnetic dipole moment
=====
***

The dipole field is generated by a series of delta function events. The recurrence time for the source events is governed by a Poisson process. The average recurrence time is fixed but the amplitude of the event is allowed to vary with a uniform distribution $U(a,b)$.  Negative (destructive) events are permitted by adopting a negative lower limit (a<0). The generation of the dipole field depends on the amplitude $A$ and the strength of the dipole field $x$.  In addition, a white noise source is added to the equations for $x$ to allow the field to reverse. We also need an additional white noise source to adjust the duration of transition fields. The imposed noise source has a zero mean and a variance $q_p$.

```julia
using Plots
using Distributions
using StatsBase
using DelimitedFiles
using DataFrames
using StatsPlots
using Printf
using Interpolations
using JSON
```

The event times are drawn from an expoential distribution with an average recurrence time

```julia
# evaluate event times
myr = 1.0e6;
recurrence = 3000/myr;         # average recurrence time (Myr)
nevents = 100000;
tevent = rand(Exponential(recurrence),nevents);  # Poisson event times
source = zeros(nevents);
relax = zeros(2*nevents);

# show distribution of event times
println("mean recurrence time = ",mean(tevent)," Myr")
histogram(tevent,label="event times",xlabel="event time (Myr)",ylabel="Number")
```

Model parameters. We adjust the values of $a$ and $b$ in the uniform distribution $U(a,b)$ to produce desired mean and variance of x(t))

```julia
# amplitude fluctuations
A = 5

amp_p = rand(Uniform(-0.4*A,5.9*A),nevents) # draw amplitude fluctuations from U(a,b)

# include random (white) noise
np = rand(Normal(),2*nevents);
qp = 180;       #  poloidal noise variance

# quenching magnitude
x0 = 4.7;         # defines onset of quenching; we chose a value slightly below the average <x>
```

```julia
function reversalFrequency(x,t)
    countReversals=0
    for i in 1:length(x)-1
        if x[i]*x[i+1] < 0
                countReversals+=1
        end
    end
    reversalRate = countReversals/t[end] #note t is usually in ma, so this returns reversals per ma
end
```

```julia
# Interpolate onto a uniform time grid

function interpolateTime(xd,td, dt=0.001)
    tinp = (0.0 : dt : td[end]);
    nodes = (td,);
    itp = interpolate(nodes,xd,Gridded(Linear()));
    xinp = itp(tinp);
    ndata = length(tinp)
    return xinp, tinp
    println(ndata)
    idecimate =1
    plot(tinp[1:idecimate:end],xinp[1:idecimate:end],
        xlabel="time",
        ylabel="dipole moment",
        label="x")
end 
```

```julia
movingaverage(g, n) = [i < n ? mean(g[begin:i]) : mean(g[i-n+1:i]) for i in 1:length(g)]
```

```julia
##Results
dfSummary = DataFrame(
    a=Float16[], b=Float16[], meanX=Float16[], stdX=Float16[], 
    skewX=Float16[], expectedX=Float16[], x0=Float16[], countReversals=Float16[], reversalRate=Float16[])

dfData = DataFrame()
dfDataInp = DataFrame()


# include random (white) noise
np = rand(Normal(),2*nevents);
qp = 180;       #  poloidal noise variance

function runModel(a,b, runNumber)
    runNumber+=1
        amp_p = rand(Uniform(a,b),nevents) # draw amplitude fluctuations from U(a,b)
        
        # quenching magnitude
        #x0 = 0.9*(a+b)/2;         # defines onset of quenching; we chose a value slightly below the average <x>
        #x0square = x0^2
        x0 = 5.0; #fix quenching at value used in published model, 
            #as shown this value actually has a big effect on the results
        
        # allocate memory for solution
        x = zeros(2*nevents);
        t = zeros(2*nevents);
        

        # set decay rate for dipole (here gamma = 1/tau_d, where tau_d is dipole decay time)
        gamma = 0.1 * 1000;
        
        expectedX = sqrt(abs.(-1*(x0^2*((a+b)/2*(1/(exp(-gamma*mean(tevent))-1))-1))))
        println(expectedX)
        
        # initial condition
        t[1] = 0.0;
        x[1] = 5.0;
        j = 2;

        # integrate over the number of source events
        for i = j : nevents # for i between j and nevents
            # relax half a time step 
            dt = tevent[i]/2.0; #tevent is Poisson distribution of times of dipole spikes
            t[j] = t[j-1] + dt;
            gp = exp(-x[j-1]^2/2.2^2);
            x[j] = x[j-1] * exp(-gamma * dt) + np[j]*sqrt(gp*qp*dt);;

            # relax rest of time step
            j += 1;
            t[j] = t[j-1] + dt;
            x[j] = x[j-1] * exp(-gamma * dt)
            relax[j] = x[j] - x[j-1]
            # evaluate quenching
            fp = 1.0/ (1.0 + (x[j]/x0)^2);

            # add event and random noise
            gp = exp(-x[j]^2/2.2^2)
            x[j] += amp_p[i] * x[j] * fp + np[j]*sqrt(gp*qp*dt); 
            source[i] = amp_p[i] * x[j] * fp + np[j]*sqrt(gp*qp*dt); #note this is negative if x is negative
            
            j+= 1;
            

        end

        # fill last entry
        t[end] = t[end-1] + tevent[end]/2.0
        x[end] = x[end-1]
        println("mean x = ",mean(abs.(x))," std x = ",std(abs.(x))," skewness = ",skewness(abs.(x)))

        meanX = mean(abs.(x))
        stdX = std(abs.(x))
        skewX = skewness(abs.(x))
        colName="t$runNumber"
        dfData[!,colName] = t
        colName="x$runNumber"
        dfData[!,colName] = x
        
        #Interpolate data on to 1kyr spaced points
        dt=0.001;
        tinp = (0.0 : dt : t[end]);
        nodes = (t,);
        itp = interpolate(nodes,x,Gridded(Linear()));
        xinp = itp(tinp);
        ndata = length(tinp)
        
        colName="t$runNumber"
        dfDataInp[!,colName] = tinp
        colName="x$runNumber"
        dfDataInp[!,colName] = xinp
        
        xinpAv = movingaverage(xinp, 25) #smooth xinp on 25kyr timescale 
        colName="xAv$runNumber"
        dfDataInp[!,colName] = xinpAv
    
        #calculate reversal rate
        countReversals=0
        for i in 1:length(xinpAv)-1
            if xinpAv[i]*xinpAv[i+1] < 0
                countReversals+=1
                #print(countReversals, "cR")
            end
        end
        print(countReversals, "cR")
        print(t[end])
        reversalRate = countReversals/t[end] #note t is usually in ma, so this returns reversals per ma
        
        xinpAv = movingaverage(xinp, 25)
        xinpAv
        push!(dfSummary, (a,b,meanX, stdX, skewX, expectedX, x0, countReversals, reversalRate))
        meanX, stdX, reversalRate
end
```

```julia
function cost(dataMean, dataStd, modelMean, modelStd)
    alpha = 1
    beta = 1
    alpha*(modelMean - dataMean)^2 + beta*(modelStd - dataStd)^2
end
```

```julia
dfCheck = DataFrame(
    runNumber=Float16[], a=Float16[], b=Float16[], modelMean=Float16[], modelStd = Float16[], costdata=Float16[])

dfFinal = DataFrame(
    runNumber=Float16[], a=Float16[], b=Float16[], modelMean=Float16[], dataMean=Float16[], modelStd = Float16[],
    dataStd=Float16[], reversalRate=Float16[], dataReversalFreq=Float16[])

#for timeperiod in [1, 2, 3, 4, 5, 6, 7]
for timeperiod in [1]
    PINTdata = Dict()
    filename = "PINTdata_" * string(timeperiod) * ".json"
    open(filename, "r") do f
    global PINTdata
    PINTdata = JSON.parse(f)  # file information to string
    end
    
    dataMean = PINTdata["mean"]
    dataStd = PINTdata["std"]
    dataReversalFreq = PINTdata["Reversal rate"]
    a=-5
    b=5
    inc = 0.1
    
    for runNumber in collect(0:1:10)
        results = runModel(a, b, runNumber)
        modelMean = results[1]
        modelStd = results[2]
        costdata = cost(dataMean, dataStd, modelMean, modelStd)

        results = runModel(a+inc, b+inc, runNumber)
        modelMean = results[1]
        modelStd = results[2]
        costPlus = cost(dataMean, dataStd, modelMean, modelStd)

        results = runModel(a-inc, b-inc, runNumber)
        modelMean = results[1]
        modelStd = results[2]
        costMinus = cost(PINTdata["mean"], PINTdata["std"], modelMean, modelStd)

        if costMinus < costPlus
            a-=inc
            b-=inc
        else
            a+=inc
            b+=inc
        end

        results = runModel(a-inc, b+inc, runNumber)
        modelMean = results[1]
        modelStd = results[2]
        costSpread = cost(dataMean, dataStd, modelMean, modelStd)

        results = runModel(a+inc, b-inc, runNumber)
        modelMean = results[1]
        modelStd = results[2]
        costNarrow = cost(dataMean, dataStd, modelMean, modelStd)

        if costSpread < costNarrow
            a-=inc
            b+=inc
        else
            a+=inc
            b-=inc
        end

        if ((costdata < costPlus && costdata < costMinus) && (costdata < costSpread && costdata < costNarrow))
            inc=inc*0.1
        end
        
        reversalRate = results[3]

        push!(dfCheck, (runNumber,a,b, modelMean, modelStd, costdata))
        println("rundone", runNumber )
        if runNumber==10
            push!(dfFinal, (runNumber, a, b, modelMean, dataMean, modelStd, dataStd, reversalRate, dataReversalFreq))
            println("last run", runNumber )
        end
        
    end
    
end
```

We integrate the governing equations for the dipole field in two steps. The event interval is divided in half. Over the first half we have dipole decay and a white noise source. Over the second half we have dipole decay, white noise and a Poisson source event. The goal is to provide intermediate solutions for $x(t)$ between the source events. This helps when we interpolate the solution onto a uniform time spacing.

```julia
##Results
dfSummary = DataFrame(
    a=Float16[], b=Float16[], meanX=Float16[], stdX=Float16[], 
    skewX=Float16[], expectedX=Float16[], x0=Float16[], countReversals=Float16[], reversalRate=Float16[])

dfData = DataFrame()
dfDataInp = DataFrame()


# include random (white) noise
np = rand(Normal(),2*nevents);
qp = 180;       #  poloidal noise variance


runNumber=0
as = collect(-1:0.1:0)
a=-1
b=5
for a in as
    bs = collect(a+1:0.1:2)
    for b in bs
    #for x0 in [1,2,3]
#for A in collect(0.1:0.1:1)
       # a=-0.4*A
        #b=5.9*A
        runNumber+=1
        amp_p = rand(Uniform(a,b),nevents) # draw amplitude fluctuations from U(a,b)
        
        # quenching magnitude
        #x0 = 0.9*(a+b)/2;         # defines onset of quenching; we chose a value slightly below the average <x>
        #x0square = x0^2
        x0 = 5.0; #fix quenching at value used in published model, 
            #as shown this value actually has a big effect on the results
        
        # allocate memory for solution
        x = zeros(2*nevents);
        t = zeros(2*nevents);
        

        # set decay rate for dipole (here gamma = 1/tau_d, where tau_d is dipole decay time)
        gamma = 0.1 * 1000;
        
        expectedX = sqrt(abs.(-1*(x0^2*((a+b)/2*(1/(exp(-gamma*mean(tevent))-1))-1))))
        println(expectedX)
        
        # initial condition
        t[1] = 0.0;
        x[1] = 5.0;
        j = 2;

        # integrate over the number of source events
        for i = j : nevents # for i between j and nevents
            # relax half a time step 
            dt = tevent[i]/2.0; #tevent is Poisson distribution of times of dipole spikes
            t[j] = t[j-1] + dt;
            gp = exp(-x[j-1]^2/2.2^2);
            x[j] = x[j-1] * exp(-gamma * dt) + np[j]*sqrt(gp*qp*dt);;

            # relax rest of time step
            j += 1;
            t[j] = t[j-1] + dt;
            x[j] = x[j-1] * exp(-gamma * dt)
            relax[j] = x[j] - x[j-1]
            # evaluate quenching
            fp = 1.0/ (1.0 + (x[j]/x0)^2);

            # add event and random noise
            gp = exp(-x[j]^2/2.2^2)
            x[j] += amp_p[i] * x[j] * fp + np[j]*sqrt(gp*qp*dt); 
            source[i] = amp_p[i] * x[j] * fp + np[j]*sqrt(gp*qp*dt); #note this is negative if x is negative
            
            j+= 1;
            

        end

        # fill last entry
        t[end] = t[end-1] + tevent[end]/2.0
        x[end] = x[end-1]
        println("mean x = ",mean(abs.(x))," std x = ",std(abs.(x))," skewness = ",skewness(abs.(x)))

        meanX = mean(abs.(x))
        stdX = std(abs.(x))
        skewX = skewness(abs.(x))
        colName="t$runNumber"
        dfData[!,colName] = t
        colName="x$runNumber"
        dfData[!,colName] = x
        
        #Interpolate data on to 1kyr spaced points
        dt=0.001;
        tinp = (0.0 : dt : t[end]);
        nodes = (t,);
        itp = interpolate(nodes,x,Gridded(Linear()));
        xinp = itp(tinp);
        ndata = length(tinp)
        
        colName="t$runNumber"
        dfDataInp[!,colName] = tinp
        colName="x$runNumber"
        dfDataInp[!,colName] = xinp
        
        xinpAv = movingaverage(xinp, 25) #smooth xinp on 25kyr timescale 
        
        #calculate reversal rate
        countReversals=0
        for i in 1:length(xinpAv)-1
            if xinpAv[i]*xinpAv[i+1] < 0
                countReversals+=1
                #print(countReversals, "cR")
            end
        end
        print(countReversals, "cR")
        print(t[end])
        reversalRate = countReversals/t[end] #note t is usually in ma, so this returns reversals per ma
        
        xinpAv = movingaverage(xinp, 25)
        xinpAv
        push!(dfSummary, (a,b,meanX, stdX, skewX, expectedX, x0, countReversals, reversalRate))
        
        
    end
end

println(dfSummary)
```

Evaluates the mean amplitudes $<A>$ and $<A^2>$, which appear in the definition of the drift $D_1$ and diffusion  $D_2$ terms.

```julia
function amplitudes(Ax,rate)
#
# evaluates the mean amplitudes <A>, <A2>, <A3.
# for poloidal and toroidal events
#
# input
# Ax, Ay - nominal amplitude of x and y events
# rate -  dimensionless rate of Poisson process

# output
# Ap, At  - mean amplitudes e.g. Ap[1,2,3]

# maximum and minimum amplitudes
Apmin = -0.4 * Ax;    # need to match the amplitude limits in U(a,b) from above
Apmax = 5.9 * Ax;



Ap = zeros(3);

Ap[1] = rate*(Apmin+Apmax)/2.0;
Ap[2] = rate*(Apmax^2 + Apmin^2 + Apmax*Apmin)/3.0;
Ap[3] = rate*(Apmax + Apmin)*(Apmax^2+Apmin^2)/4.0;



return Ap

end
```

```julia
# compute drift and diffusion terms
rate = 1/recurrence
xbar = mean(abs.(x));
Ap = amplitudes(A,rate)


xplot = (0 : 0.1 : 10.0);
D1 = zeros(length(xplot))
D2 = zeros(length(xplot))
v = zeros(length(xplot))

for i = 1 : length(xplot)
    f = 1.0 / (1.0 + (xplot[i]/x0)^2);
    g = exp(-xplot[i]^2/2.2^2)
    D1[i] = -gamma*xplot[i] + Ap[1]*xplot[i]*f;
    D2[i] = 0.5*Ap[2]*(xplot[i]^2)*(f^2) + 0.5 * qp*g;
   # D3[i] = (1/6.0) * Ap[3]*(xplot[i]^2)*(f^3);
end

println(maximum(D2))
plot(xplot,D1,xlabel="Dipole Moment", label = "D1")
plot!(xplot,D2,label = "D2")
```
